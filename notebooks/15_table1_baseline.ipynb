# 15 · Table 1 — Baseline characteristics (publication quality)

Publication-ready Table 1 via `gtsummary`. Two tables:
- **Table 1a.** Full PPMI cohort (N=1,485) at baseline.
- **Table 1b.** Matched cohort (N=170) at baseline.

Groups: DBS vs Never-DBS. Standardized mean differences reported. Save HTML, PNG, and docx-compatible formats.

```r
source("helpers/pain_helpers.R")
suppressPackageStartupMessages({
  library(dplyr); library(tidyr); library(gtsummary); library(gt)
})

vars <- c("age_at_visit","ageonset","duration_yrs","SEX","BMI","LEDD",
          "updrs3_score","NHY","NP1PAIN","NP1SLPN","NP1SLPD","NP1FATG",
          "NP1URIN","NP1DPRS","NP1ANXS","gds","stai","ess","rem","scopa")
labels_list <- list(
  age_at_visit ~ "Age at baseline (yrs)",
  ageonset     ~ "Age at PD onset (yrs)",
  duration_yrs ~ "Disease duration (yrs)",
  SEX          ~ "Sex (Male=1)",
  BMI          ~ "Body-mass index",
  LEDD         ~ "Levodopa-equivalent daily dose (mg)",
  updrs3_score ~ "MDS-UPDRS-III motor score",
  NHY          ~ "Hoehn & Yahr",
  NP1PAIN      ~ "NP1PAIN (0–4)",
  NP1SLPN      ~ "NP1SLPN (sleep problems)",
  NP1SLPD      ~ "NP1SLPD (daytime sleepiness)",
  NP1FATG      ~ "NP1FATG (fatigue)",
  NP1URIN      ~ "NP1URIN (urinary)",
  NP1DPRS      ~ "NP1DPRS (depression)",
  NP1ANXS      ~ "NP1ANXS (anxiety)",
  gds          ~ "Geriatric Depression Scale",
  stai         ~ "STAI (state + trait)",
  ess          ~ "Epworth Sleepiness Scale",
  rem          ~ "RBD questionnaire",
  scopa        ~ "SCOPA-AUT (autonomic)"
)
```

```r
# Matched cohort (notebook 00 derived object)
df <- readRDS(file.path(OUT_OBJ, "pain_long.rds"))
base_m <- df %>% dplyr::arrange(PATNO, INFODT_orig) %>%
  dplyr::group_by(PATNO) %>% dplyr::slice_head(n = 1) %>% dplyr::ungroup() %>%
  dplyr::select(PATNO, will_receive_dbs, dplyr::any_of(vars)) %>%
  dplyr::mutate(arm = dplyr::if_else(will_receive_dbs, "DBS", "Never-DBS"))

t1_matched <- base_m %>% dplyr::select(-PATNO, -will_receive_dbs) %>%
  gtsummary::tbl_summary(
    by = arm,
    missing = "no",
    type = list(gtsummary::all_continuous() ~ "continuous2",
                gtsummary::all_categorical() ~ "categorical"),
    statistic = list(gtsummary::all_continuous() ~ c("{mean} ({sd})", "{median} ({p25}, {p75})")),
    label = labels_list
  ) %>%
  gtsummary::add_p() %>% gtsummary::add_overall() %>%
  gtsummary::modify_caption("**Table 1a. Baseline characteristics of the matched cohort (N=170).**")

gtsummary::as_gt(t1_matched) %>%
  gt::gtsave(filename = file.path(OUT_TAB, "Table1a_matched.html"))

t1_matched
```

```r
# Full cohort (patient-anchor frame — one baseline per patient)
rel <- load_full_ppmi_rel_patient_anchor()
base_f <- rel %>% dplyr::filter(months >= -24, months <= 0) %>%
  dplyr::arrange(PATNO, dplyr::desc(months)) %>%
  dplyr::group_by(PATNO) %>% dplyr::slice_head(n = 1) %>% dplyr::ungroup() %>%
  dplyr::select(PATNO, will_receive_dbs, dplyr::any_of(vars)) %>%
  dplyr::mutate(arm = dplyr::if_else(will_receive_dbs, "DBS", "Never-DBS"))
cat("Full-cohort baseline n:", nrow(base_f), "\n")
print(base_f %>% dplyr::count(arm))

t1_full <- base_f %>% dplyr::select(-PATNO, -will_receive_dbs) %>%
  gtsummary::tbl_summary(
    by = arm, missing = "no",
    type = list(gtsummary::all_continuous() ~ "continuous2"),
    statistic = list(gtsummary::all_continuous() ~ c("{mean} ({sd})", "{median} ({p25}, {p75})")),
    label = labels_list
  ) %>%
  gtsummary::add_p() %>% gtsummary::add_overall() %>%
  gtsummary::modify_caption("**Table 1b. Baseline characteristics of the full PPMI cohort (N≈1,484).**")

gtsummary::as_gt(t1_full) %>%
  gt::gtsave(filename = file.path(OUT_TAB, "Table1b_full.html"))
t1_full
```

```r
save_object(base_m, "table1_matched_data")
save_object(base_f, "table1_full_data")
cat("Saved.\n")
```
